```python
import cv2
print(cv2.__version__)
```

```python
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("CORRPT.jpg")
```

```python
if image is None:
    print("Error:Image not found")
else:
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    edges = cv2.Canny(blurred, 50, 150)
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    axs[0, 0].imshow(image_rgb)
    axs[0, 0].set_title("Money Original Image")
    axs[0, 1].imshow(gray, cmap="gray")
    axs[0, 1].set_title("Money Grayscale Image")
    axs[1, 0].imshow(blurred, cmap="gray")
    axs[1, 0].set_title("Money Blurred Image")
    axs[1, 1].imshow(edges, cmap="gray")
    axs[1, 1].set_title("Money Edge Detected Image")
for ax in axs.flat:
    ax.axis("off")
plt.show()


```
